### ðŸŽ¯ Guess the Number Game ðŸ’­ðŸŽ²âœ¨
This is a simple Python game where the player has to guess a randomly chosen mystery number!

```python
import random 
# Used Python's built-in module to generate a random mystery number for this game

# ==== GLOBAL COUNTERS ====
# This tracks the Player's game statistics across multiple plays
try:
    total_games = 0 
except NameError: 
    total_games = 0
    total_wins = 0
    total_losses = 0
```

```python
def choose_difficulty(): # Asking the Player to choose a difficulty level to play and return(low, high, max_trie) for that level
    
    print("Choose a difficulty level:")
    print("1 - Easy (1 - 10, 3 tries)")
    print("2 - Medium (1 - 20, 4 tries)")
    print("3 - Hard (1 - 50, 5 tries)")
    print("4 - god-mode (1 - 100, 7 tries)")

    # Loop until the Player enters a valid menu option
    while True:
        choice = input("Enter 1, 2, 3, or 4: ").strip()
        # return settings based on the selected difficulty
        if choice == "1":
            return 1, 10, 3
        elif choice == "2":
            return 1, 20, 4
        elif choice == "3":
            return 1, 50, 5
        elif choice == "4":
            return 1, 100, 7
        else: 
            # any input other than 1-4 is invalid; ask the Player to enter again.
            print("Invalid choice. Please enter 1, 2, 3, or 4.")
```

```python
def play_game(name):
     # access the global counter variables declared above
    global total_games, total_wins, total_losses
    total_games += 1  # increment total games played every time this function runs

    low, high, max_tries = choose_difficulty() # configure game from prior difficulty selection 
    
    # low = minimum possible mystery number
    # high = maximum possible mystery number
    # max_tries = how many guesses the Player gets
    
    mystery_number = random.randint(low, high) # Player picks a random secret number in the configured range (inclusive of 1 and max. number in range)

    print(f"\nHello {name}, I'm thinking of a number between {low} and {high}.")
    print(f"You have {max_tries} attempts to guess the mystery number.\n")

    for guesses_taken in range(1, max_tries + 1):
        # input validation loop: keep asking Player until a valid integer is entered
        while True:
            try:
                # f-string shows the current attempt number
                user_guess = int(input(f"Attempt {guesses_taken}: What's your guess? ðŸ’­ >> "))
                break
            except ValueError:
                # where a non-number is entered, prompt Player again without losing their attempt count
                print("Please enter a valid number!")

        if user_guess < mystery_number:
            print("Your guess is too low :/ Try again!\n")
        elif user_guess > mystery_number: 
            print("Your guess is too high :/ Try again!\n")
        else:
            total_wins += 1
            # Correct guess: Player celebrates and exits the attempts loop early
            print(f"ðŸŽ‰ðŸŽ‰ Well done, {name}! You guessed the correct number in {guesses_taken} tries!")
            break
    else: 
        total_losses += 1
        # this 'else' belongs to the for-loop
        # it will only run if the loop didn't 'break' (e.g., the player never guessed correctly)
        print(f"Boo ðŸ‘Ž You lose! ðŸ‘Ž The correct mystery number was {mystery_number}. Better luck next time {name}!")

    # ==== Show Player's current Game Statistics ====
    print("\nðŸ“ˆðŸ“ŠðŸ“‰ Game Statistics:")
    print(f"Total games played: {total_games}")
    print(f"Wins: {total_wins}")
    print(f"Losses: {total_losses}\n")
```

```python
def keep_playing():
    # Main game loop controller
    # Repeats the game until the Player chooses to stop

    name = input("Enter your name before we start: ").strip()  # stores name once
    
    while True: 
        play_game(name) # runs one full round of the game, includes Player's name
        
        while True:
            again = input(f"{name}, do you want to play again?: ").strip().lower() # asks the Player if they want to play again
            if again in ("yes", "no"): # anything other than 'yes' ends the game loop
                break 
            print("Please enter 'y' for yes or 'n' for no.")

        if again.startswith("y"):
            continue
        else:
            print(f"Thanks for playing, {name}! ðŸ‘‹")
            break  # Exits the loop and ends the function
```

```python
keep_playing()
```
